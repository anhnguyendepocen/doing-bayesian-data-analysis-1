```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
from scipy.stats import beta as dbeta
from bern_beta import bern_beta
from dbda2e_utilities import hdi_of_icdf
%matplotlib inline
```

```python
flips = Series([1, 1, 1, 1, 1, 1, 1, 1, 1, 0])
```

## (A)

```python
govt_prior = (50, 50)
govt_posterior = bern_beta(govt_prior, flips)
```

```python
govt_post_beta = dbeta(govt_posterior[0], govt_posterior[1])
```

```python
hdi_of_icdf(govt_post_beta)
```

```python
govt_post_beta.mean()
```

The government minted coin has a strong prior of being fair, the new 10 flips don't change it much.

Theres a 95% chance that the probability of Heads lies in the range $[0.44, 0.63]$ 

Since these flips are independent, the mean of the posterior is a good choice for the probability of a 11th heads and 
is 0.54

## (B)

```python
magic_prior = (0.1, 0.1)
magic_posterior = bern_beta(magic_prior, flips)
```

```python
magic_post_beta = dbeta(magic_posterior[0], magic_posterior[1])
```

```python
hdi_of_icdf(magic_post_beta)
```

```python
magic_post_beta.mean()
```

The Magic coin is probably biased towards Heads or Tails but I don't know which. After 9 heads, the posterior shifts heavily in favour of a Heads biased coin.

Theres a 95% chance the probability of Heads lies in the range $[0.65, 0.99]$

Since these flips are independent, the mean of the posterior is a good choice for the probability of a 11th heads and 
is 0.89
