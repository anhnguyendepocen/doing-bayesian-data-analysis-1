```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
from bern_beta import bern_beta
from dbda2e_utilities import hdi_of_icdf
from scipy.stats import beta as dbeta
%matplotlib inline
```

## (A)

```python
prior = (1, 1)
```

```python
newspaper_poll = Series(np.concatenate([np.repeat(1, 58),
                                        np.repeat(0, 100-58)]))
newspaper_posterior = bern_beta(prior, newspaper_poll)
```

```python
hdi_of_icdf(dbeta(newspaper_posterior[0], newspaper_posterior[1]))
```

$ [0.48, 0.67] $ covers 95% of the probability mass 

## (B)

```python
followup_poll = Series(np.concatenate([np.repeat(1, 57),
                                       np.repeat(0, 100-57)]))
followup_posterior = bern_beta(newspaper_posterior, followup_poll)
```

```python
hdi_of_icdf(dbeta(followup_posterior[0], followup_posterior[1]))
```

The bounds are now narrower with $[0.51,0.64]$ covering 95% of the probability mass
