```python
import numpy as np
from scipy.stats import beta as dbeta
from bern_metrop import metropolis, plot_trajectory
%matplotlib inline
```

```python
my_data = np.concatenate([np.repeat(0, 6), np.repeat(1, 14)])
```

```python
def prior(theta):
    return dbeta(1, 1).pdf(theta)

def likelihood(theta):
    N = len(my_data)
    z = sum(my_data)
    p_data_1_theta = theta**z * (1-theta)**(N-z)
    return p_data_1_theta
```

## Metropolis with SD=0.02

```python
trajectory = metropolis(prior, likelihood, proposal_sd=0.02)
plot_trajectory(trajectory)
```

## Metropolis with SD=0.2

```python
trajectory = metropolis(prior, likelihood, proposal_sd=0.2)
plot_trajectory(trajectory)
```

## Metropolis with SD=2.0

```python
trajectory = metropolis(prior, likelihood, proposal_sd=2.0)
plot_trajectory(trajectory)
```

The 3 plots appear to be converging but have very different acceptance ratios.

A Metropolis run with a SD of 0.02 takes small steps and explores very little of the distribution with an ess of only 555

A Metropolis run with a SD of 2.0 takes large steps that have low likelihood and are rarely accepted, resulting in an unrepresentative posterior wih an ess of 1341

A Metropolis run with a SD of 0.2 has the best ess by a wide margin (8086) and is the most representative.
