```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
from bern_beta import bern_beta
from dbda2e_utilities import hdi_of_icdf
from scipy.stats import beta as dbeta
%matplotlib inline
```

```python
prior = (1, 1)
```

Taking $F$ as 1 and $J$ as 0

```python
radio = Series(np.concatenate([np.repeat(1, 40),
                               np.repeat(0, 10)]))
ocnmnt = Series(np.concatenate([np.repeat(1, 15),
                                np.repeat(0, 35)]))
```

## Radio HDI

```python
radio_posterior = bern_beta(prior, radio)
```

```python
hdi_of_icdf(dbeta(radio_posterior[0], radio_posterior[1]))
```

Theres a 95% chance that the probability of choosing $F$ lies in the range $[0.67, 0.89]$. This indicates a bias towards $F$ 

## OcnMnt HDI

```python
ocnmnt_posterior = bern_beta(prior, ocnmnt)
```

```python
hdi_of_icdf(dbeta(ocnmnt_posterior[0], ocnmnt_posterior[1]))
```

Theres a 95% chance that the probability of choosing $F$ lies in the range $[0.19, 0.44]$. This indicates a bias towards $J$
