```python
import numpy as np
import matplotlib.pyplot as plt
from bern_metrop import metropolis, plot_trajectory
%matplotlib inline
```

## (A)

```python
def prior(theta):
    return ((np.cos(4*np.pi*theta) + 1) ** 2)/1.5
```

## (B)

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.set_title("prior")
ax.set_xlabel("$\\theta$")
ax.set_ylabel("$\\frac{cos(4\\cdot\\pi\\cdot\\theta)^2}{1.5}$")
ax.plot(np.arange(0, 1, 1.0/500), prior(np.arange(0, 1, 1.0/500)))
```

## (C)

```python
def blank_likelihood(theta):
    p_data_1_theta = 1
    return p_data_1_theta


def two_by_three_likelihood(theta):
    N = 3
    z = 2
    p_data_1_theta = theta**z * (1-theta)**(N-z)
    return p_data_1_theta
```

### Blank Likelihood

```python
trajectory = metropolis(prior, blank_likelihood)
plot_trajectory(trajectory)
```

The posterior histogram looks a little like the prior

## (D)

### $\frac{2}{3}$ likelihood

```python
trajectory = metropolis(prior, two_by_three_likelihood)
plot_trajectory(trajectory)
```

The posterior histogram is skewed to the right

## (E)

### Blank likelihood, sd=0.02

```python
trajectory = metropolis(prior, blank_likelihood, proposal_sd=0.02)
plot_trajectory(trajectory)
```

This posterior histogram recreates the left 1/3rd of the prior quite accurately but nothing else

### $\frac{2}{3}$ likelihood, sd=0.02

```python
trajectory = metropolis(prior, two_by_three_likelihood, proposal_sd=0.02)
plot_trajectory(trajectory)
```

This posterior histogram ignores the high probability of tails

## (F)

### Blank likelihood, sd=0.02, start=0.99

```python
trajectory = metropolis(prior, blank_likelihood, proposal_sd=0.02, start_position=0.99)
plot_trajectory(trajectory)
```

This posterior histogram recreates the right 1/3rd of the prior quite accurately but nothing else

### $\frac{2}{3}$ likelihood, sd=0.02, start=0.99

```python
trajectory = metropolis(prior, two_by_three_likelihood, proposal_sd=0.02, start_position=0.99)
plot_trajectory(trajectory)
```

This posterior histogram, skews a bit left to take into account the 1/3rd tail chance
