```python
import numpy as np
from scipy.stats import beta as dbeta
from bern_metrop import metropolis, plot_autocorrelation
%matplotlib inline
```

```python
my_data = np.concatenate([np.repeat(0, 6), np.repeat(1, 14)])
```

```python
def prior(theta):
    return dbeta(1, 1).pdf(theta)

def likelihood(theta):
    N = len(my_data)
    z = sum(my_data)
    p_data_1_theta = theta**z * (1-theta)**(N-z)
    return p_data_1_theta
```

## Metropolis with SD=0.02

```python
trajectory = metropolis(prior, likelihood, proposal_sd=0.02)
plot_autocorrelation(trajectory)
```

## Metropolis with SD=0.2

```python
trajectory = metropolis(prior, likelihood, proposal_sd=0.2)
plot_autocorrelation(trajectory)
```

## Metropolis with SD=2.0

```python
trajectory = metropolis(prior, likelihood, proposal_sd=2.0)
plot_autocorrelation(trajectory)
```
